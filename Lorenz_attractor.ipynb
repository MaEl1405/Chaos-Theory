<h1 style="color: blue;">Lorenz Attractor & Chaos Theory </h1>

The Lorenz attractor is a strange attractor living in 3D space that relates three parameters arising in fluid dynamics. It is one of the Chaos theory's most iconic images and illustrates the phenomenon now known as the Butterfly effect or (more technically) sensitive dependence on initial conditions.

In the following, we implement the Lorenz attractor in Python and investigate the effect of small changes in the initial conditions on the output



```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Equations
def lorenz(t, xyz, sigma, rho, beta):
    x, y, z = xyz
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]
```

```python
#Parameters
sigma = 10
rho = 28
beta = 8/3

#Initial values
xyz0_list = [
    [0.1, 0, 0],
    [0.101,0, 0],
]

#time span
t_eval = np.linspace(0, 100, 10000)
```

```python
#solving differential Equations
sol = solve_ivp(lorenz, [0, 100], xyz0_list[1], args=(sigma,rho,beta), dense_output=True)
x, y, z = sol.sol(t_eval)
plt.figure()
plt.plot(x, z)
plt.title('x,z')

plt.figure()
plt.plot(x,y)
plt.title('x,y')

plt.figure()
plt.plot(y,z)
plt.title('y,z')

plt.show()
```

```python
fig = plt.figure()
for i, xyz0 in enumerate(xyz0_list):
    sol = solve_ivp(lorenz, [0, 100], xyz0, args=(sigma,rho,beta), dense_output=True)

    #Visualization
    x, y, z = sol.sol(t_eval)
    plt.subplot(3,1,1)
    plt.plot(x)
    plt.title('x')
    
    plt.subplot(3,1,2)
    plt.plot(y)
    plt.title('y')
    
    plt.subplot(3,1,3)
    plt.plot(z)
    plt.title('z')
    plt.subplots_adjust(hspace=0.6)
    

plt.subplots_adjust(hspace=0.3)
plt.show()


```
